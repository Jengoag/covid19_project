```python
# creamos diccionario para poder tener los datos organizados y poder crear un df con tres columnas 
recovered_dict = recovered.to_dict("records")
recovered_dict
```

```python
# Creamos un nuevo dataframe, este tendrá tres columnas Country, Date, Ocurrence.

df = pd.DataFrame(columns=['Country', 'Date', 'Occurrence'])

recovered_dict = recovered.to_dict("records")
for i in recovered_dict:
    country = i['Country']
    for key in i:
        if key != "Country" and "/22/" in key or "/8/" in key:
           df = df.append({'Country': country, 'Date': key, 'Occurrence': i[key]}, ignore_index=True)

df 
```

```python
#creamos la base de datos en sql

df.to_sql("recovered", engine)

print("Created table recovered in Postgres")
```
